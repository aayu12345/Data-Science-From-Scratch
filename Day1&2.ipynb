### 1) Variables & Data Types
**Definition:** A variable is a container used to store data in Python.
Python is *dynamically typed* — you don't need to declare the type.
Common data types:
- `int`, `float`
- `str` (string)
- `bool`
- `list`, `tuple`, `set`, `dict`


```python
x=5
pi=3.4
name='Ayush Srivastava'
is_active=True
print(type(x),x)
print(type(pi),pi)
print(type(name),name)
print(type(is_active),is_active)
```

### 2) Type Casting

Convert between types using `int()`, `float()`, `str()`, `list()` etc.


```python
# Type casting examples
s = '123'            # string

# nu1 = int(s)        # string to int
# nu1 = in
num = int(s)        # string to int

f = float('3.5')    # string to float
print(num, type(num))
print(f, type(f))

# Implicit casting example
result = x + pi
print('x + pi =', result, type(result))
```

### 3) ➗ Operators (Arithmetic, Logical, Comparison)

Arithmetic, comparison, and logical operators are frequently used in data processing.


```python
a = 8
b = 3
print('a + b =', a + b)
print('a - b =', a - b)
print('a * b =', a * b)
print('a / b =', a / b)

print('a // b =', a // b)  # floor division
print('a % b =', a % b)    # remainder

# Comparison
print('a > b ->', a > b)
print('a == b ->', a == b)

# Logical
print('a > 5 and b < 5 ->', a > 5 and b < 5)
print('not (a < 5) ->', not (a < 5))

```
